```python
import pandas as pd
from docx2python import docx2python
from itertools import chain
```

```python
doc = docx2python("dataset/kaz_tarih_2016.docx")
```

```python
doc.body[0][0][0]
```

```python
result = []
iter = 0
for i in range(1,len(doc.body[0][0][0])):
    if(len(doc.body[0][0][0][i]) > 3):
        if(doc.body[0][0][0][i][1] != 't'):
            result.append(doc.body[0][0][0][i])
            iter +=1
        else:
            result[iter] = result[iter] + " " + doc.body[0][0][0][i] + " " + doc.body[0][0][0][i + 1]
```
