```python
import pandas as pd
import numpy as np
from docx2python import docx2python
from itertools import chain

import re
```

```python
def preprocess(raw_text):
    raw_text = raw_text.replace('\xa0', ' ')
    raw_text = raw_text.replace('\t', ' ')
    raw_text = raw_text.replace('.', ' ')
    raw_text = re.sub('\s+', ' ', raw_text)
    
    return raw_text
```

```python
docd = docx2python("dataset/tarih/tarih_duman_1.docx")
print(len(docd.body[0][0][0]))
```

```python
doca = docx2python("dataset/tarih/tarih_asel_1.docx")
print(len(doca.body[0][0][0]))
```

```python
doca.body[0][0][0][16835:]
```

```python
lstd = doca.body[0][0][0][:15999] + docd.body[0][0][0]
ls = []
for i in range(len(lstd)):
    aa = preprocess(lstd[i])
    if(aa != '' and aa != ' ' and aa != '  ' and aa != '   ' and len(aa.split(" ")) > 0):
        ls.append(aa)
        
print(len(ls))
```

```python
pd.set_option('display.max_rows', 10)
df = pd.DataFrame(np.array(ls[:]).reshape(-1,2).tolist(),
                  columns = ['question', 'answer'])

df
```

```python
df.to_csv("dataset/tarih.csv", index = False)
df
```

```python
def find_answer(que, ans):
    result = []
    for i in range(que.shape[0]):
        if(ans.Answer[i] == 'A'):
            result.append([que.Question[i][3:].lower(), que.A[i][3:].lower()])
        elif(ans.Answer[i] == 'B'):
            result.append([que.Question[i][3:].lower(), que.B[i][3:].lower()])
        elif(ans.Answer[i] == 'C'):
            result.append([que.Question[i][3:].lower(), que.C[i][3:].lower()])
        elif(ans.Answer[i] == 'D'):
            result.append([que.Question[i][3:].lower(), que.D[i][3:].lower()])
        elif(ans.Answer[i] == 'E'):
            result.append([que.Question[i][3:].lower(), que.E[i][3:].lower()])
    
    return pd.DataFrame(result, columns = ['question', 'answer'])
```
